Rule-based approach

```python
import re
import string
import pandas as pd

# Define rules for classifying tweets
positive_keywords = ['bom', 'grande', 'maravilha', 'amor']
negative_keywords = ['ruim', 'terrivel', 'odio', 'detesto']
#edu_keywords = ['educar', 'escola', 'ensino', 'colegio', 'Educação', 'educação', 'alunos', 'professores', 'colégio']
#keywords = ['saúde', 'hospital', 'médico', 'vacina', 'Dengue', 'remédio', 'saude', 'medico', 'remedio']
emp_keywords = ['emprego', 'empregos', 'salario', 'salário', 'renda', 'vagas', 'trabalho', 'trabalhos', 'trabalhador']

# Function to classify tweets based on rules
def classify_tweet(tweet):
    # Convert tweet to lowercase
    tweet = tweet.lower()
    # Check if tweet contains positive keywords
    if any(keyword in tweet for keyword in positive_keywords):
        return 'positive'
    # Check if tweet contains negative keywords
    elif any(keyword in tweet for keyword in negative_keywords):
        return 'negative'
    # Check if tweet contains spam keywords
    elif any(keyword in tweet for keyword in sau_keywords):
        return 'emp'
    # If no rules match, classify as neutral
    else:
        return 'neutral'
```

```python
df_airline['rulebased_class'] = df_airline['string_content'].apply(classify_tweet)

df_airline[['Content','rulebased_class']]
```
